```python
# analyse des données
# etude sur la production des voitures electriques en fonction au temps
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid") # grille blanc au fond du graphique
# adaptation des voitures électrique en fonction du temps
plt.figure(figsize=(12,6))
adoption_par_an = data['Model Year'].value_counts().sort_index()
sns.barplot(x= adoption_par_an.index,y=adoption_par_an.values)
plt.title('adaptation des voitures au fur du temps')
plt.xlabel('Model Year')
plt.ylabel('Nombre de véhicules enregistré')
plt.show()
```

```python
# Etude du nombre de véhicules enregistré dans les villes des 5 departements principaux(county)

county_distribution = data['County'].value_counts()
top_counties = county_distribution.head(5).index
# Filtrage du Jeu de Données pour Ces Principaux Comtés
top_counties_data = data[data['County'].isin(top_counties)]
# Analyse de la Distribution des VE au Sein des Villes de Ces Principaux Comtés
city_distribution_top_counties = top_counties_data.groupby(['County', 'City']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')
# Visualisation dans les 10 principaux ville de chaque compté
top_cities = city_distribution_top_counties.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='Number of Vehicles', y='City', hue='County', data=top_cities, palette="magma")
plt.title('Top Cities in Top Counties by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('City')
plt.legend(title='County')
plt.tight_layout()
plt.show()
```

```python
#Analyse de la répartition des différentes type de vihicules
type_distribution = data['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x= type_distribution.values, y= type_distribution.index)
plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Electric Vehicle Type')
plt.tight_layout()
plt.show()
```

```python
#analyse des constructeurs Les plus populaire sur le marché
make_distribution = data['Make'].value_counts().head(10) 
plt.figure(figsize=(12, 6))
sns.barplot(x= make_distribution.values, y= make_distribution.index)
plt.title('Top 10 Popular EV Makes')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Make')
plt.tight_layout()
plt.show()
```

```python
#Sélection des 3 Principaux Constructeurs Basée sur le Nombre de Véhicules Enregistrés
top_3_makes = make_distribution.head(3).index
#Filtrage du Jeu de Données pour Ces Principaux Constructeurs
top_makes_data = data[data['Make'].isin(top_3_makes)]

#Analyse de la Popularité des Modèles de VE au Sein de Ces Principaux Constructeurs
model_distribution_top_makes = top_makes_data.groupby(['Make', 'Model']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')
#Visualisation des 10 Principaux Modèles de Ces Constructeurs pour Plus de Clarté
top_models = model_distribution_top_makes.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='Number of Vehicles', y='Model', hue='Make', data=top_models, palette="viridis")
plt.title('Top Models in Top 3 Makes by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.tight_layout()
plt.show()
```

```python
average_range_by_model = top_makes_data.groupby(['Make', 'Model'])['Electric Range'].mean().sort_values(ascending=False).reset_index()

# the top 10 models with the highest average electric range
top_range_models = average_range_by_model.head(10)

plt.figure(figsize=(12, 8))
barplot = sns.barplot(x='Electric Range', y='Model', hue='Make', data=top_range_models, palette="cool")
plt.title('Top 10 Models by Average Electric Range in Top Makes')
plt.xlabel('Average Electric Range (miles)')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.show()
```

```python
# Calcul du Nombre de Véhicules Électriques Enregistrés Chaque Année
registration_counts = data['Model Year'].value_counts().sort_index()
registration_counts
```

```python
from scipy.optimize import curve_fit
import numpy as np
```

```python
#Prévision du Nombre de Véhicules Électriques Enregistrés dans les années prochaines

# Filtrer les données pour n'inclure que les années pour lesquelles les données sont complètes, jusqu'à 2023.
filtered_years = registration_counts[registration_counts.index <= 2023]
```

```python
# Définir une fonction de croissance exponentielle pour ajuster les données.
def exp_growth(x, a, b):
    return a * np.exp(b * x)
#Préparer les Données pour l'Ajustement de Courbe
x_data = filtered_years.index - filtered_years.index.min() #centrer les années en anée minimal
y_data = filtered_years.values
#Ajuster les Données à la Fonction de Croissance Exponentielle
params, covariance = curve_fit(exp_growth, x_data, y_data)
# Prévoir le Nombre de Véhicules Électriques pour 2024 et les Cinq Années Suivantes
forecast_years = np.arange(2024, 2024 + 6) - filtered_years.index.min()
forecasted_values = exp_growth(forecast_years, *params)
#Créer un Dictionnaire pour Afficher les Valeurs Prévisionnelles
forecasted_evs = dict(zip(forecast_years + filtered_years.index.min(), forecasted_values))
print(forecasted_evs)
```
